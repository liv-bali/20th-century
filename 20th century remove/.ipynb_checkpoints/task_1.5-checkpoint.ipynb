## 1. Import libraries and txt files

```python
from textblob import TextBlob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import nltk
import nltk
import re
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from collections import Counter
sns.set()
```

## 2. Tokenization

```python
myfile = open('20th_century_wiki.txt', encoding='utf-8') 
```

```python
# Import txt file

with open('20th_century_wiki.txt', 'r', errors='ignore') as file:
    data = file.read().replace('\n', '')
```

```python
# Sentence tokenization 

from nltk.tokenize import sent_tokenize
tokenized_sent = sent_tokenize(data)
print(tokenized_sent)
```

```python
# Word tokenization

from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(data)
print(tokenized_word)
```

```python
# Create frequency distribution

from nltk.probability import FreqDist
dist_words = FreqDist(tokenized_word)
print(dist_words)
```

```python
dist_words.most_common(10)
```

```python
# Frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words.plot(10,cumulative = False)
plt.show()
```

## 3. Stopwords

```python
# Defining stopwords

from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
print(stop_words)
```

```python
# Removing stopwords in sentences

# filtered_sent = []
# for word in tokenized_sent:
#     if word not in stop_words:
#         filtered_sent.append(word)
# print("Tokenized Sentence:", tokenized_sent)
# print("Filterd Sentence:", filtered_sent)
```

```python
# Removing stopwords in words

filtered_words = [] # creates an empty list
for word in tokenized_word:
    if word not in stop_words:
        filtered_words.append(word)
```

```python
filtered_words
```

```python
# Create a new FreqDist for filtered_words

dist_words_filter = FreqDist(filtered_words)
print(dist_words_filter)
```

```python
# Frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words_filter.plot(10, cumulative = False)
plt.show()
```

```python
dist_words_filter
```

```python
# Substitute all punctuations marks with a space 

sans_punc = re.sub("[^a-zA-Z]",  # Search for all non-letters
                          " ",          # Replace all non-letters with spaces
                          str(filtered_words))
```

```python
sans_punc
```

```python
# Word tokenization

tokenized_word_2 = word_tokenize(sans_punc)
print(tokenized_word_2)
```

```python
# Create a new FreqDist

dist_words_filter_2 = FreqDist(tokenized_word_2)
```

```python
# Frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words_filter_2.plot(30, cumulative = False)
plt.show()
```

```python
dist_words_filter_2.most_common(20)
```

## 4. POS tags with TextBlob

```python
new_stopwords = ["And", "Then", 'n', 't', 's', 'The']
```

```python
filtered = []
for word in tokenized_word_2:
    if word not in new_stopwords:
        filtered.append(word)
```

```python
%%time
text = TextBlob(str(filtered))
```

```python
text
```

```python
tags_list = text.tags
```

```python
tags_list
```

```python
df_text = pd.DataFrame(tags_list)
df_text.columns = ['Words', "Word type"]
```

```python
df_text.head()
```

```python
df_t = df_text.groupby('Word type').count().reset_index()
```

```python
df_t.head()
```

```python
top20 = df_t.nlargest(20, 'Words')
```

```python
plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:blue", 20):
    sns.barplot(x = "Words", y = "Word type",
    saturation = 0.9, data = top20).set_title("20th Century Histroy Wiki - top 20 word types used")
```

## 5. Nouns

```python
df = df_text[(df_text['Word type'] == "NN") | (df_text['Word type'] == "NNS") | (df_text['Word type'] == "NNP")]
df.columns = ["Word", "Occurences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurences'], ascending=False)
top10 = y.nlargest(10, 'Occurences')
```

```python
top10
```

```python
plt.figure(figsize=(10, 5))
with sns.dark_palette("xkcd:blue", 10):
    sns.barplot(x="Word", y="Occurences",
    saturation=0.9, data = top10).set_title("20th Century History Wiki - most frequently used nouns")
```

## 6. Verbs

```python
df = df_text[(df_text['Word type'] == "VB")  | (df_text['Word type'] == "VBD")]
df.columns = ["Word", "Occurences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurences'], ascending=False)
top10 = y.nlargest(10, 'Occurences')
```

```python
top10
```

```python
plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:blue", 10):
    sns.barplot(x = "Word", y = "Occurences",
    saturation = 0.9, data = top10).set_title("20th Century History Wiki - most frequently used verbs")
```

## 7. Adjectives

```python
df = df_text[df_text['Word type'] == "JJ"]
df.columns = ["Word", "Occurences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by=['Occurences'], ascending=False)
top10 = y.nlargest(10, 'Occurences')
```

```python
plt.figure(figsize=(10, 5))
with sns.dark_palette("xkcd:blue", 10):
    sns.barplot(x="Word", y="Occurences",
    saturation=0.9, data=top10).set_title("20th Century History Wiki - most frequently used adjectives")
```

## 8. Counting country occurences

```python
listToStr = ' '.join([str(elem) for elem in filtered])
 
print(listToStr)
```

```python
# Create a count for countries

all_counts = Counter(re.sub(r'\W+', ' ', listToStr).split())
```

```python
import geopandas as gpd

# Load the world dataset (update the path to your file)
world = gpd.read_file("C:/Users/olivi/OneDrive/Documents/Data Analytics/Data Visualizations with Python/ne_110m_admin_0_countries")
# Extract country names/ne_110m_admin_0_countries.shp")

# Extract country names
countries = world['NAME'].tolist()

print(countries)
```

```python
# Convert to DataFrame
df_countries = pd.DataFrame(countries, columns=["Country"])
```

```python
df_countries.head()
```

```python
country_list = df_countries['Country'].to_list()
```

```python
dict_of_counts = {d : all_counts[d] for d in country_list}
```

```python
dict_of_counts
```

```python
# Search for the names from the list in the dictionary

dct = {v:[k] for v,k in dict_of_counts.items()}  
df = pd.DataFrame(dct)
```

```python
df
```

```python
df = df.transpose().reset_index()
```

```python
df.dtypes
```

```python
df
```

```python
df.rename(columns = {"index":"Country", 0:"Times mentioned"}, inplace = True)
```

```python
df
```

```python
df.shape
```

```python
plt.figure(figsize=(12, 6))  # Adjusted figure size for clarity

# Sort the DataFrame and keep only the top 50 countries
top_countries = df.sort_values("Times mentioned", ascending=False).head(40)

# Create the barplot
sns.barplot(x="Times mentioned", y="Country", data=top_countries, saturation=0.9, palette="Blues_r")

# Set the title
plt.title("20th Century History Wiki - Top 40 Most Frequently Mentioned Countries")

# Show the plot
plt.show()
```
